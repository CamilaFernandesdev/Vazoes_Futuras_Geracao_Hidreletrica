## TESTE QUI-QUADRADO

[Acervo Lima - Teste QUI-quadrada](https://acervolima.com/python-teste-qui-quadrado-de-pearson/)
A distribuição χ2 ou qui-quadrado é uma das distribuições mais utilizadas em estatística inferencial, principalmente para realizar testes de χ2. Este teste serve para avaliar quantitativamente a relação entre o resultado de um experimento e a distribuição esperada para o fenômeno. Isto é, ele nos diz com quanta certeza os valores observados podem ser aceitos como regidos pela teoria em questão. Muitos outros testes de hipótese usam, também, a distribuição χ2.

A função chi2_contingency() do módulo scipy.stats toma como entrada a tabela de contingência no formato de array 2d. Ele retorna uma tupla contendo estatísticas de teste , o valor p , graus de liberdade e a tabela esperada (aquela que criamos a partir dos valores calculados) nessa ordem. 

Por isso, é necessário comparar o obtido p-valor com alfa valor de 0,05.

Podemos verificar a hipótese por estes métodos:

Usando o valor p :
Definimos um fator de significância para determinar se a relação entre as variáveis ​​é de significância considerável. Geralmente, um fator de significância ou valor alfa de 0,05 é escolhido. Este valor alfa denota a probabilidade de rejeitar H0 erroneamente quando for verdadeiro. Um valor alfa inferior é escolhido nos casos em que esperamos mais precisão. Se o valor p para o teste for estritamente maior do que o valor alfa, H0 será verdadeiro.

Usando o valor qui-quadrado :
Se nosso valor calculado de qui-quadrado for menor ou igual ao valor tabular (também chamado de crítico ) de qui-quadrado, então H0 será verdadeiro.

```python
import scipy
from scipy.stats impor chi2_contingency

# defining the table
data = [[207, 282, 241], [234, 242, 232]]
stat, p, dof, expected = chi2_contingency(data)
  
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```
